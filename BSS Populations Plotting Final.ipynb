```python
%matplotlib inline
pd.set_option('display.max_columns', None)

data=unpickle(path='C:/Users/GG/Dropbox/uni/HonsProject/Age 8-10/data/tuto_data_sin.pkl')

cmd=unpickle(path='C:/Users/GG/Downloads/cmd_030v2.pckl')

age_arr = []
frac_arr = []
blue_num = []
red_num = []
bss_age_list = []
bss_files_list=[]
for i in range(78,101,1):
    i /= 10
    age = i
    age_bin = (age-6)/0.1
    age_bin=int(age_bin)
    weight = BPASS_TIME_INTERVALS[int(age_bin)]
    
    data['B-V'] = (data['B']-data['V'])   #Create column for color B-V
    data['log_age'] = np.log10(data['age'].replace(0, np.nan))
    #data_age=data[data['M1']==1.2]
    data_age=data[np.round(data['log_age'],1)==age]
    data_age = data_age.sort_values('B-V')
    
    col_WD, mag_WD = (data_age[(data_age['V'] > 2) & (data_age['B-V'] < -0.15)].sort_values('V'))['B-V'], (data_age[(data_age['V'] > 2) & (data_age['B-V'] < -0.15)].sort_values('V'))['V']
    
    
    if age <= 9.1:
        col_MSTO, mag_MSTO = data_age[data_age['V'] < 1]['B-V'], data_age[data_age['V'] < 1]['V']
        col_RGB, mag_RGB = (data_age[(data_age['B-V'] > (col_MSTO.iloc[0]+0.2)) & (data_age['V'] < 1)].sort_values('V', ascending = False))['B-V']-0.1,(data_age[(data_age['B-V'] > (col_MSTO.iloc[0]+0.2)) & (data_age['V'] < 1)].sort_values('V', ascending = False))['V']
    elif age <=9.7 and age>9.1:
        col_MSTO, mag_MSTO = data_age[data_age['V'] < 3]['B-V'], data_age[data_age['V'] < 3]['V']
        col_RGB, mag_RGB = (data_age[(data_age['B-V'] > (col_MSTO.iloc[0]+0.2)) & (data_age['V'] < 3)].sort_values('V', ascending = False))['B-V'],(data_age[(data_age['B-V'] > (col_MSTO.iloc[0]+0.2)) & (data_age['V'] < 3)].sort_values('V', ascending = False))['V']
    else:
        col_MSTO, mag_MSTO = data_age[(data_age['V'] < 4) & (data_age['B-V'] > 0)]['B-V'], data_age[(data_age['V'] < 4) & (data_age['B-V'] > 0)]['V']
        col_RGB, mag_RGB = (data_age[(data_age['B-V'] > (col_MSTO.iloc[0]+0.2)) & (data_age['V'] < 4)].sort_values('V', ascending = False))['B-V'],(data_age[(data_age['B-V'] > (col_MSTO.iloc[0]+0.2)) & (data_age['V'] < 4)].sort_values('V', ascending = False))['V']    
    fig, ax = plt.subplots(1,2, figsize=(9,4))
    myplot = cmd.plot(loc=121, log_age=age) # Here you can chose the time bin you want to plot.
    
    #ax[0].scatter(col_MSTO.iloc[0],mag_MSTO.iloc[0], color = 'white', marker='.')
    #try:
    #    ax[0].scatter(col_WD.iloc[0], mag_WD.iloc[0], color = 'red', marker='.')
    #except Exception:
    #    pass
    
    #ax[0].scatter(col_RGB.iloc[0], mag_RGB.iloc[0], color = 'blue', marker='.')
    
    myplot.set_xlabel('B-V')
    myplot.set_ylabel('V')
    myplot.set_xlim([-1,2.0])
    myplot.set_ylim([6.5,-8])
    myplot.set_title('M67 z=0.03 age={}'.format(age))
    
    myplot_sin = cmd_sin.plot(loc=122, log_age=age)
    
    myplot_sin.set_xlabel('B-V')
    myplot_sin.set_ylabel('V')
    myplot_sin.set_xlim([-1,2.0])
    myplot_sin.set_ylim([6.5,-8])
    myplot_sin.set_title('M67 z=0.03 age={}'.format(age))
    
    XB=[]
    for mag in range(len(cmd.mag_range)):
        for col in range(len(cmd.col_range)):
            if cmd_sin.grid[age_bin, mag, col]/weight != min(cmd_sin.grid[age_bin][0])/weight and cmd.grid[age_bin, mag, col]/weight != min(cmd.grid[age_bin][0])/weight and cmd_sin.col_range[col]<=(col_MSTO.iloc[0]):
                XB.append((cmd_sin.col_range[col],cmd_sin.mag_range[mag]))
    #            ax[1].scatter(cmd_sin.col_range[col], cmd_sin.mag_range[mag])

    closest_index = distance.cdist([(col_MSTO.iloc[0], mag_MSTO.iloc[0])], XB).argmin()
    #ax[1].scatter(XB[closest_index][0], XB[closest_index][1], marker='x')
    
    if age <= 9.1:
        RGB_start = 0.66#col_MSTO.iloc[0]+0.5
    else:
        RGB_start = 0.66 #col_MSTO.iloc[0]+0.3
        
    BSS_rect = patches.Rectangle((col_MSTO.iloc[0]-0.05,mag_MSTO.iloc[0]+1), width=-20, height=-20, linewidth=1, edgecolor='b', facecolor='none', linestyle = '--')
    RGB_rect = patches.Rectangle((col_RGB.iloc[0],mag_RGB.iloc[0]), width=20, height=-20, linewidth=1, edgecolor='r', facecolor='none', linestyle = '--')
    ax[0].add_patch(BSS_rect)
    #ax[0].scatter(col_MSTO.iloc[0],mag_WD.iloc[0], marker='o', c='green')
    ax[0].add_patch(RGB_rect)
    BSS_rect = patches.Rectangle((col_MSTO.iloc[0]-0.05,mag_MSTO.iloc[0]+1), width=-20, height=-20, linewidth=1, edgecolor='b', facecolor='none', linestyle = '--')
    RGB_rect = patches.Rectangle((col_RGB.iloc[0],mag_RGB.iloc[0]), width=20, height=-20, linewidth=1, edgecolor='r', facecolor='none', linestyle = '--')
    ax[1].add_patch(BSS_rect)
    ax[1].add_patch(RGB_rect)
    
    
    #PLOT SCATTER
    #for i in range(len(m67_obs)):
    #    if (m67_obs.iloc[i]['B-V'] < (col_MSTO.iloc[0])-0.07) and (m67_obs.iloc[i]['V']<(mag_MSTO.iloc[0]+1-0.07)):
    #        ax[0].scatter(m67_obs.iloc[i]['B-V'], m67_obs.iloc[i]['V'], color='royalblue', s=2)
    #    else:
    #        ax[0].scatter(m67_obs.iloc[i]['B-V'], m67_obs.iloc[i]['V'], color='green', s=2)
    
    
    
    #GET BSS FRACTIONS
    BSS_sum = 0
    for mag in range(len(cmd.mag_range)):
        for col in range(len(cmd.col_range)):
            if cmd.col_range[col] < (col_MSTO.iloc[0]) and cmd.mag_range[mag]<(mag_MSTO.iloc[0]+1) and cmd.grid[age_bin, mag, col]/weight != cmd.grid[age_bin,0,0]/weight and cmd_sin.grid[age_bin, mag, col]/weight == min(cmd_sin.grid[age_bin][0])/weight:
                BSS_sum += cmd.grid[age_bin, mag, col]/weight
                bss_files_list.append(cmd.files[age_bin, mag, col])
                bss_age_list.append(age)
                #ax[0].scatter(cmd.col_range[col], cmd.mag_range[mag], c='red')
        
    RGB_sum = 0
    for mag in range(len(cmd.mag_range)):
            for col in range(len(cmd.col_range)):
                if cmd.col_range[col] > col_RGB.iloc[0] and cmd.mag_range[mag]<(mag_RGB.iloc[0]) and cmd.grid[age_bin, mag, col]/weight != cmd.grid[age_bin,0,0]/weight:
                    RGB_sum += cmd.grid[age_bin, mag, col]/weight
    
    
    plt.figure()
    sizes=[BSS_sum, RGB_sum] 
    
    plt.pie(sizes, colors=['blue', 'red'])
    plt.title('Model')
    print(sizes)
    frac_arr.append(BSS_sum/RGB_sum)
    age_arr.append(i)
    blue_num.append(sizes[0])
    red_num.append(sizes[1])
```
